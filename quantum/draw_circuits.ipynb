```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit 
from qiskit import execute, transpile, Aer, IBMQ
from qiskit.visualization import plot_histogram
import math
import matplotlib
```

```python
draw_kwargs = {
    "output": "mpl",
    "cregbundle": False,
    "initial_state": True,
    "idle_wires": False,
    "style": {
        "name": "bw",
        "subfontsize": 9,
        "dpi": 600
} }
histogram_color = "#82caaf"
# use matplotlib
# separate classical register wires
# show |0> and 0
# don't show unused wires
# black-and-white for book
# font size of subscripts
# image resolution
```

```python
qreg_q = QuantumRegister(2, "q") 
creg_c = ClassicalRegister(2, "meas")
circuit = QuantumCircuit(qreg_q, creg_c)
circuit.x(qreg_q)
circuit.draw(**draw_kwargs)
```

```python
circuit.measure(qreg_q[0], creg_c[0])
circuit.draw(**draw_kwargs)
```

```python
simulator = Aer.get_backend("aer_simulator")
```

```python
result = execute(circuit, simulator, shots=1000).result() 
counts = result.get_counts(circuit)
counts
```

```python
histogram = plot_histogram(counts, color=histogram_color, title="IBM Quantum Simulator")
histogram
```
